## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate. 

```python
theta_0=0
theta_1=0
alpha=0.001
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
y_predict=theta_0+theta_1*x
print(y_predict)
```

### Step3: Calculate Cost function 𝑱(theta_0,theta_1 ).


```python
cost=0
for i in range(len(x)):
    cost+=(y_predict[i]-y[i])**2
cost/=(2*len(x))
print(cost)
```

### Step4: Calculate the gradient.


```python
Grad_theta0=0
Grad_theta1=0

for i in range(len(x)):
    Grad_theta0+=(y_predict[i] - y[i])
    Grad_theta1+=(y_predict[i] - y[i])*x[i]
    
Grad_theta0/=len(x)
Grad_theta1 /=len(x)
print(Grad_theta0)
print(Grad_theta1)
```

### Step5: Update the parameters (simultaneously).


```python
theta_0=theta_0- alpha * Grad_theta0
theta_1=theta_1-alpha*Grad_theta1
print(theta_0)
print(theta_1)
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.

```python
#Batch Gradient Descent for Single variable Linear regression
def Batch_Gradient_Descent(X,Y):
    # 1.  Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.
    theta_0=0
    theta_1=0
    alpha=0.001
    
    #Loop for 100 epochs
    for e in range(100):
        print("==============================================================")
        print("epoch ",e)
        #2. Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
       
        Y_predict=[]
        for i in range (len(X)):
            Y_predict.append(theta_0+theta_1 *X[i])

        #3. Calculate Cost function 𝑱(theta_0,theta_1 ).
        cost=0
        for i in range(len(X)):
            cost+=(Y_predict[i] - Y[i])**2
        cost /= (2*len(X))

        print("Cost= " , cost)
        #4.  Calculate gradient of theta_0 and theta_1

        Grad_theta_0=0
        Grad_theta_1=0
        for i in range(len(X)):

            Grad_theta_0 += Y_predict[i] - Y[i]
            Grad_theta_1 += (Y_predict[i] - Y[i])*X[i]
        Grad_theta_0 /=(len(X))
        Grad_theta_1 /= (len(X))

        #5. Update the parameters (simultaneously).
        theta_0 =theta_0 -alpha*Grad_theta_0

        theta_1= theta_1 -alpha *Grad_theta_1


    return Y_predict , theta_0 ,theta_1
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
y_hat,th0,th1=Batch_Gradient_Descent(x,y)
```

```python
print(y_hat)
```

#### Plot  LR equation output (fitted line) with the original data (scatter plot of X,y)

```python
r2_score(y,y_hat)
```

## GD vectorize Implementation
### Implement GD without iterate through data points i.e. use vector operations

```python
#vectorized implementation to Batch gradient Descent

loss=[]
th0_list=[]
th1_list=[]
def vectorized_Batch_Gradient_Descent(X,Y):
    # 1.  Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.
    theta_0=0
    theta_1=0
    alpha=0.001
    
    #Loop for 100 epochs
    for e in range(100):
        print("==============================================================")
        print("epoch ",e)
        
        th0_list.append(theta_0)
        th1_list.append(theta_1)
        
        #2. Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
       
        Y_predict=theta_0 + theta_1 *X
        #3. Calculate Cost function 𝑱(theta_0,theta_1 ).
        error=Y_predict - Y
        
        cost= (np.dot(error,error))/(2*len(X))
        loss.append(cost)

        print("Cost= " , cost)
        #4.  Calculate gradient of theta_0 and theta_1

        Grad_theta_0=(Y_predict - Y).sum()/(len(X))
        Grad_theta_1=(np.dot((Y_predict-Y),X) )/(len(X))

        #5. Update the parameters (simultaneously).
        theta_0 =theta_0 -alpha*Grad_theta_0

        theta_1= theta_1 -alpha *Grad_theta_1


    return Y_predict , theta_0 ,theta_1
    

```

### Repeat your implementation but for more than one variable

```python
loss=[]

def multivariate_GD(X,Y):
    """
    X is a matrix contain the multi features
    Y is a vector contains actual data
    """
    Y=Y.reshape(Y.shape[0],1)
    #1.assume theta vector,alpha
    theta_vector=np.zeros(X.shape[1]).reshape(X.shape[1],1)
    thetas=theta_vector.T
    alpha=0.0001
   
    for e in range(100):
        
        print("=======================================")
        print("epoch " ,e)
        
        
        #2. calculate Y_predict
        Y_predict=np.dot(X,theta_vector)
    
        #3.calculate cost function
    
        cost=0
        error=(Y_predict-Y)
        cost=np.dot(error.T,error)/(2*X.shape[0])
        loss.append(cost[0][0])
        print("cost= ",cost[0][0])
        
        #4.calculate gradient vector
        Grad_vector=np.zeros(X.shape[1])
        
        Grad_vector=np.dot( X.T,(Y_predict - Y))/(X.shape[0])
        
        #5.update theta
        theta_vector=theta_vector - alpha * Grad_vector
        thetas=np.vstack([thetas,theta_vector.T])
        
        
    return Y_predict , theta_vector,thetas
    
    
    
```
